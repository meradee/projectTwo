```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
```

### ***Diversity Rate Dataset***

```python
diversitycsv = "Resources/diversityindex.csv"
diversity_df = pd.read_csv(diversitycsv)
diversity_df.head()
```

```python
# Separating County and State 
cleaned_df = diversity_df
cleaned_df['County'] = diversity_df['Location'].str[:-4] 
cleaned_df['State'] = diversity_df['Location'].str[-2:]
cleaned_df.head() 
```

```python
#Extracting only State CA
ca_df = cleaned_df.loc[cleaned_df['State'] == 'CA']
ca_df.head()
```

```python
#Reset index
cleaned_ca_df = ca_df.reset_index(drop=True)
cleaned_ca_df.head()
```

```python
#Transform: Calculating the averages 
mean_cleaned_ca_df = pd.DataFrame({"Diversity-Index":cleaned_ca_df.groupby(["State","County"])["Diversity-Index"].mean(),
                                   "Black":cleaned_ca_df.groupby(['State','County'])['Black or African American alone, percent, 2013'].mean(),
                                   "American Indian":cleaned_ca_df.groupby(["State","County"])["American Indian and Alaska Native alone, percent, 2013"].mean(),
                                   "Asian":cleaned_ca_df.groupby(["State","County"])["Asian alone, percent, 2013"].mean(),
                                   "Native Hawaiian":cleaned_ca_df.groupby(["State","County"])["Native Hawaiian and Other Pacific Islander alone, percent,"].mean(),
                                   "Two or More Races":cleaned_ca_df.groupby(["State","County"])["Two or More Races, percent, 2013"].mean(),
                                   "Hispanic/ Latino":cleaned_ca_df.groupby(["State","County"])["Hispanic or Latino, percent, 2013"].mean(),
                                   "White":cleaned_ca_df.groupby(["State","County"])["White alone, not Hispanic or Latino, percent, 2013"].mean()
                                  })
mean_cleaned_ca_df.head(10)
```

```python
mean_cleaned_ca_df.plot.bar(legend=True, rot=90)
plt.ylabel('Number of Data Points')
plt.title('Data Points Per County in CA')
plt.gcf().set_size_inches(20, 10)
plt.tight_layout()
plt.savefig('Barplot_Diversity_CA_County')
plt.show()
```

### ***Unemployment Rate Dataset***

```python
unemployment_df = pd.read_csv('Resources/unemployment_output.csv')
unemployment_df.head()
```

```python
us_state_abbrev = {'California': 'CA'}
```

```python
for key, value in us_state_abbrev.items():
    unemployment_df.loc[unemployment_df['State']==key,'State'] = value 
```

```python
ca_unemployment_df = unemployment_df.loc[unemployment_df['State'] == 'CA'] 
ca_unemployment_df.head()
```

```python
ca_unemployment_df['State'].count()
```

```python
#Note: float = can be decimal
ca_unemployment_df.dtypes
```

```python
#Extracting the data of years 2013 only
year_unemployment_df = ca_unemployment_df.loc[ca_unemployment_df['Year'] == 2013]
year_unemployment_df.head()
```

```python
mean_unemployment_df = pd.DataFrame({"Unemployment Rate 2013":year_unemployment_df.groupby(['State','County'])['Rate'].mean()})
mean_unemployment_df.head()
```

```python
mean_unemployment_df.plot.bar(legend=True, rot=90)
plt.ylabel('Percentage of Data Points')
plt.title('Unemployment Data Percentage Per County in CA')
plt.gcf().set_size_inches(20, 10)
plt.tight_layout()
plt.savefig('Barplot_Unemployment_CA_County')
plt.show()
```

# Merge Data

```python
merged_df = pd.merge(mean_cleaned_ca_df,mean_unemployment_df, on=["State",'County'],how="inner")
```
